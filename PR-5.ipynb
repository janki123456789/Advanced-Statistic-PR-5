Norm-1 (Manhattan Norm) Formula :
 
 

```python
norm1_A = np.linalg.norm(A, 1)
norm1_B = np.linalg.norm(B, 1)
norm1_C = np.linalg.norm(C, 1)

print("Norm-1 of A:", norm1_A)
print("Norm-1 of B:", norm1_B)
print("Norm-1 of C:", norm1_C)
```

Norm-2 (Euclidean Norm) formula :




```python
norm2_A = np.linalg.norm(A)
norm2_B = np.linalg.norm(B)

print("Norm-2 of A:", norm2_A)
print("Norm-2 of B:", norm2_B)
```

#### (B)

Dot Product formula :
 


```python
dot_product = np.dot(A, B)

print("Dot Product:", dot_product)
```

Angle Between Two Vectors formula:



Finally :



```python
cos_theta = dot_product / (norm2_A * norm2_B)
angle = math.degrees(math.acos(cos_theta))

print("Angle between A and B:", angle)
```

```python
cross_product = np.cross(A, B)

print("Cross Product:", cross_product)
```

### Line, Plane, and Hyperplane with respect to Dataset Dimensions

In linear algebra and geometry, the dimensionality of data determines how it is represented in space.

1. Line (1D)

A line represents one-dimensional space. If our dataset contains only one feature 
(for example, only Math scores), each student’s score can be represented as a 
point on a single line.

Example:
Student Score = [85]

Here the dataset has only one dimension, so the data lies on a line.


2. Plane (2D)

A plane represents two-dimensional space. If the dataset contains two subjects 
such as Math and Physics, each student’s scores can be represented as a point 
in a 2D coordinate system.

Example:
Student Score = [Math, Physics] = [85, 80]

Each student becomes a point on a plane with coordinates (x, y).


3. 3D Space

When three subjects are considered (Math, Physics, Chemistry), the dataset 
becomes three-dimensional. Each student’s scores form a point in 3D space.

Example:
Student Score = [Math, Physics, Chemistry] = [85, 80, 78]

Each student is represented as a point (x, y, z).


4. Hyperplane (Higher Dimensions)

A hyperplane is a generalization of a plane in higher-dimensional spaces. 
If the dataset contains more than three subjects (for example Math, Physics, 
Chemistry, Biology, English), the data exists in higher dimensions such as 
4D or 5D.

In such cases, hyperplanes are used to represent or separate data in 
higher-dimensional space.

Summary:

1 Subject → 1D → Line  
2 Subjects → 2D → Plane  
3 Subjects → 3D → Space  
n Subjects → n-Dimensional Space → Hyperplane

### Dimensionality Increase from 2D → 3D → Higher Dimensions with Hyperplanes

Dimensionality means the number of features (subjects) used to represent data.
In the student performance dataset, each subject represents one dimension.

2D Space (Plane) :

If we consider two subjects such as Math and Physics, the dataset becomes
two-dimensional. Each student is represented as a point on a plane.

Example:
Student Score = (Math, Physics) = (85, 80)

In 2D space, the hyperplane is a line that can separate the data.

3D Space :

When we add another subject such as Chemistry, the dataset becomes
three-dimensional.

Example:
Student Score = (Math, Physics, Chemistry) = (85, 80, 78)

In 3D space, the hyperplane becomes a plane that can divide the dataset.


Higher Dimensions

If more subjects are added (for example Biology, English, Computer Science),
the dataset moves into higher dimensions such as 4D, 5D, or n-dimensional space.

These higher-dimensional spaces cannot be visualized easily, but mathematically
they exist and are represented using hyperplanes.

A hyperplane is a generalization of a line or plane that separates data points
in higher-dimensional space.


Summary

2D → Hyperplane is a line  
3D → Hyperplane is a plane  
n-D → Hyperplane separates higher-dimensional data

Thus, as the number of features (subjects) increases, the dimensionality of the
dataset also increases from 2D to 3D and then to higher dimensions represented
by hyperplanes.

- Eigenvectors = directions where transformation acts as scaling only (no rotation/skew).

- Eigenvalues = how much scaling happens along those directions.

Eigenvalues Formula :



Eigenvectors Formula :



```python
data = {
    "Math": [85, 78, 92],
    "Physics": [80, 75, 88],
    "Chemistry": [78, 82, 85]
}

df = pd.DataFrame(data)
matrix = df.values
df
```

```python
cov_matrix = np.cov(matrix, rowvar=False)

print("Covariance Matrix:")
print(cov_matrix)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:")
print(eigenvalues)

print("Eigenvectors:")
print(eigenvectors)
```

Singular Value Decomposition (SVD) decomposes a matrix into three matrices:

A = U Σ Vᵀ

U → Left singular vectors  
Σ → Singular values  
Vᵀ → Right singular vectors  

Role in Dimensionality Reduction:

- SVD helps reduce the number of dimensions by keeping only the largest
- singular values and removing smaller ones. This preserves the most
- important information in the dataset while reducing complexity.

Use : 

- Recommender systems(Netflix , Spotify)
- Image compression
